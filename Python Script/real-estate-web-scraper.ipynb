```python
dataset_1 = pd.read_csv("../data/Island_Luxury_Data.csv")
dataset_2 = pd.read_csv("../data/Lekki_Central_Data.csv")
dataset_3 = pd.read_csv("../data/ajah_properties.csv")
```

```python
dataset_1.columns = ["Title", "Location", "Price", "Bedrooms", "Bathrooms", "Toilets", "Property Type", "URL"]
dataset_2.columns = ["Title", "Location", "Price", "Bedrooms", "Bathrooms", "Toilets", "Property Type", "URL"]
dataset_3 = dataset_3.drop(columns=dataset_3.columns[0])
dataset_3.columns = ["Title", "Location", "Price", "Bedrooms", "Bathrooms", "Toilets", "Property Type", "URL"]

merged_dataset = pd.concat([dataset_1, dataset_2, dataset_3])
merged_dataset
```

```python
merged_dataset["Bedrooms"] = merged_dataset["Bedrooms"].astype(str)
merged_dataset = merged_dataset[~ (merged_dataset["Bedrooms"].str.contains("Toilet", case=False))]
merged_dataset = merged_dataset[~ (merged_dataset["Bedrooms"].str.contains("Toilets", case=False))]
merged_dataset = merged_dataset[~ (merged_dataset["Bedrooms"].str.contains("Bathroom", case=False))]
merged_dataset = merged_dataset[~ (merged_dataset["Bedrooms"].str.contains("Bathrooms", case=False))]
merged_dataset["Bedrooms"] = merged_dataset["Bedrooms"].str.replace("s", "").str.strip()
merged_dataset["Bedrooms"] = merged_dataset["Bedrooms"].str.replace("S", "").str.strip()
merged_dataset
```

```python
merged_dataset["Bathrooms"] = merged_dataset["Bathrooms"].astype(str)
merged_dataset["Bathrooms"] = merged_dataset["Bathrooms"].str.replace("Bathroom", "").str.replace("Bathrooms", "").str.strip()
merged_dataset = merged_dataset[~ (merged_dataset["Bathrooms"].str.contains("Toilet", case=False))]
merged_dataset = merged_dataset[~ (merged_dataset["Bathrooms"].str.contains("Toilets", case=False))]
merged_dataset = merged_dataset[~ (merged_dataset["Bathrooms"].str.contains("Bedrooms", case=False))]
merged_dataset = merged_dataset[~ (merged_dataset["Bathrooms"].str.contains("Bedroom", case=False))]
merged_dataset["Bathrooms"] = merged_dataset["Bathrooms"].str.replace("s", "").str.strip()
merged_dataset["Bathrooms"] = merged_dataset["Bathrooms"].str.replace("S", "").str.strip()

merged_dataset["Toilets"] = merged_dataset["Toilets"].astype(str)
merged_dataset["Toilets"] = merged_dataset["Toilets"].str.replace("Toilet", "").str.replace("Toilets", "").str.strip()
merged_dataset = merged_dataset[~ (merged_dataset["Toilets"].str.contains("Bedroom", case=False))]
merged_dataset = merged_dataset[~ (merged_dataset["Toilets"].str.contains("Bedrooms", case=False))]
merged_dataset = merged_dataset[~ (merged_dataset["Toilets"].str.contains("Bathroom", case=False))]
merged_dataset = merged_dataset[~ (merged_dataset["Toilets"].str.contains("Bathrooms", case=False))]
merged_dataset["Toilets"] = merged_dataset["Toilets"].str.replace("s", "").str.strip()
merged_dataset["Toilets"] = merged_dataset["Toilets"].str.replace("S", "").str.strip()

merged_dataset
```

```python
merged_dataset["Bathrooms"] = merged_dataset["Bathrooms"].astype(int)
merged_dataset["Bedrooms"] = merged_dataset["Bedrooms"].astype(int)
merged_dataset["Toilets"] = merged_dataset["Toilets"].astype(int)

merged_dataset.dropna(subset = ["Bathrooms"], inplace = True)
merged_dataset.dropna(subset = ["Toilets"], inplace = True)

merged_dataset
```

```python
merged_dataset.to_csv("../data/Cleaned_Merged_Dataset.csv", index=False)
```
